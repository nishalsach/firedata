```python
import numpy as np
import pandas as pd
from IPython.display import display 

df = pd.read_csv(filepath_or_buffer='/Users/sachita/Documents/Projects/Summer/Winter2018/train_data.txt', 
                 delimiter='\t') 
```

```python
df_elim = df.loc[:, ['date', 'lat', 'lon', 'cld', 'gfed', 'npp', 'pr', 'prev_ba', 'rh', 'ts', 'ftmap0',
 'ftmap1',
 'ftmap2',
 'ftmap3',
 'ftmap4',
 'ftmap5',
 'ftmap6',
 'ftmap7',
 'ftmap8',
 'ftmap9',
 'ftmap10', 'pop']]

df_elim.head()
```

```python
#Convert to log
#df_elim['class'] = df['gfed'].apply(lambda x: 0 if x==0 else 1)
df_elim['log10'] = df['gfed'].apply(lambda x: np.log10(x) if x!=0 else 0)
df_elim.head()
```

```python
#Getting classes from log values

limits = np.linspace(-6, 0, 25) #list of upper bounds of each bin
labels = list(range(1, 25))
bin_labels=[str(x) for x in labels]
valuepass = df_elim['log10'].tolist()

final_labels = pd.cut(valuepass, bins=limits, labels=bin_labels, include_lowest=True)
df_elim = df_elim.assign(labels=final_labels)

def define_label(row):
    if row['log10']==0:
        return 0
    else:
        return row['labels']

df_elim['final_label'] = df_elim.apply(define_label, axis = 1)
df_elim.head()
```

```python
df = df_elim.drop(['log10', 'labels', 'prev_ba'], axis=1)
df.head()
```

```python
df.to_csv(path_or_buf='/Users/sachita/Documents/Projects/Summer/Winter2018/firedata/fire_mod.csv', sep=',', index=False)
```

```python
#Sort by lat lon and cut up into discrete datasets
df = df.sort_values(['lat', 'lon'])
df
```

```python
#14 years worth of monthly data => 168 readings/lat-lon pair
def splitDataFrameIntoSmaller(df, chunkSize): 
    listOfDf = list()
    numberChunks = len(df) // chunkSize 
    print(numberChunks)
    for i in range(numberChunks):
        listOfDf.append(df[i*chunkSize:(i+1)*chunkSize])
    return listOfDf

result = splitDataFrameIntoSmaller(df, 168)
```
